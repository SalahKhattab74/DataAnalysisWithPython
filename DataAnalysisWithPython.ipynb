Basics Of Grouping

```python
df['drive-wheels'].unique()
```

```python
df_group_one = df[['drive-wheels','body-style','price']]
```

```python
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
```

```python
df_group_one
```

```python
df_gptest = df[['drive-wheels','body-style','price']]
group_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
```

```python
group_test1
```

```python
group_pivot = group_test1.pivot(index='drive-wheels',columns='body-style')
group_pivot
```

```python
group_pivot = group_pivot.fillna(0)
```

```python
group_pivot
```

```python
df_group_two = df[['body-style','price']]
grouped_test_bodystyle = df_group_two.groupby(['body-style'],as_index=False).mean()
```

```python
grouped_test_bodystyle
```

```python
plt.pcolor(group_pivot,cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig , ax = plt.subplots()
im = ax.pcolor(group_pivot,cmap='RdBu')

#label Names 
row_labels = group_pivot.columns.levels[1]
col_labels = group_pivot.index

#move ticks and labels to the center 
ax.set_xticks(np.arange(group_pivot.shape[1])+0.5,minor = False)
ax.set_yticks(np.arange(group_pivot.shape[0])+0.5,minor = False)

#insert labels 
ax.set_xticklabels(row_labels,minor=False)
ax.set_yticklabels(col_labels,minor=False)

#rotate labe if too long 
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
from scipy import stats 
```

```python
pearson_coef , p_values = stats.pearsonr(df['wheel-base'],df['price'])
print("The Pearson Correlation Coeffiecent Is ", pearson_coef, "with a P-Value of P = ", p_values)
```

Since The P-Value is <0.001 the correlation between wheel-base and price is statistically significant , although the linear relationship isn't extremely strong 

```python
pearson_coef , p_values = stats.pearsonr(df['length'],df['price'])
print("The Pearson Correlation Coeffiecent Is ", pearson_coef, "with a P-Value of P = ", p_values)
```
